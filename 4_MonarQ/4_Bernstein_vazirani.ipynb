# Dfi rcapitulatif : l'algorithme de Bernstein-Vazirani

Vous allez maintenant mettre en pratique tout ce que vous avez appris sur la programmation quantique en relevant le dfi suivant. L'objectif est d'implmenter l'algorithme de Bernstein-Vazirani. Cet algorithme quantique permet de dcouvrir une chane binaire cache $s$ de longueur $n$ en une seule valuation de la fonction oracle $ f(x)$, l o un algorithme classique ncessiterait $n$ valuations. Cet algorithme montre un avantage quantique clair pour certains problmes de calcul.


## Contexte et Problme

L'algorithme rsout le problme suivant : tant donn une fonction boolenne $f(x)$ dfinie sur des chanes binaires $x$ de longueur $n$, la fonction $f(x)$ est linaire et de la forme :

\begin{align}
       f(x) = s \cdot x = s_0 x_0 \oplus s_1 x_1 \oplus \cdots \oplus s_{n-1} x_{n-1}
    \end{align}



o $ s = (s_0, s_1, \dots, s_{n-1}) $ est une chane binaire fixe inconnue, et $ \oplus $ reprsente l'addition modulo 2, le rsultat est donc 0 ou 1.

\begin{align}
       1 \oplus 0 = 0 \oplus 1=1
          \end{align}

\begin{align}
       1 \oplus 1 =0 \oplus 0=0
          \end{align}


**But** : Trouver la chane binaire $ s $ en une seule valuation de la fonction $ f(x) $.

## tapes de l'algorithme

### 1. Prparation des qubits

Nous commenons avec $ n $ qubits d'entre initialiss  l'tat $ |0\rangle $ et un qubit auxiliaire initialis  l'tat $ |1\rangle $. L'tat initial est donc:
\begin{align}
       |0\rangle^{\otimes n} |1\rangle   
    \end{align}


### 2. Superposition

Appliquez une porte Hadamard $ H $  chaque qubit d'entre. La porte Hadamard transforme un qubit dans l'tat $ |0\rangle $ en une superposition gale de $ |0\rangle $ et $ |1\rangle $ :

\begin{align}
        H|0\rangle = \frac{1}{\sqrt{2}} (|0\rangle + |1\rangle)    
    \end{align}


Applique  tous les qubits, cela produit l'tat superpos suivant pour les $n $ qubits d'entre:

\begin{align}
        \frac{1}{\sqrt{2^n}} \sum_{x=0}^{2^n-1} |x\rangle 
    \end{align}



Le qubit auxiliaire reste dans l'tat $|1\rangle $, transform en $ \frac{1}{\sqrt{2}} (|0\rangle - |1\rangle)$ par une porte Hadamard.

### 3. Application de l'oracle

L'oracle est une bote noire qui implmente la fonction $ f(x) $. Cette fonction est telle que $ f(x) = s \cdot x $, o $ s $ est la chane binaire inconnue que nous voulons dcouvrir. L'oracle applique une porte CNOT conditionnelle  chaque qubit d'entre en fonction de $ s $. Si le bit $ s_i $ de $ s $ est 1, alors l'oracle applique une porte CNOT contrle par le qubit d'entre $ x_i $ et cible le qubit ancillaire.

L'oracle est dfini par l'opration unitaire suivante :
\begin{align}
U_f|x\rangle|y\rangle=|x\rangle|y \oplus(f(x))\rangle
\end{align}
avec: 
* $|x\rangle$ l'input de l'oracle (la superposition des $x$-tats)
* $|y\rangle$ le qubit ancillaire dans l'tat $|-\rangle$

L'oracle inverse donc l'tat du qubit ancillaire en fonction de la valeur de $ f(x)$
\begin{align}
|-\rangle=\frac{|0\rangle-|1\rangle}{\sqrt{2}}\quad\Rightarrow\quad|y\oplus f(x)\rangle=\frac{|f(x)\rangle-|1\oplus f(x)\rangle}{\sqrt{2}}
\end{align}

En effet, si $f(x)=1$ alors: 

\begin{aligned}
|f(x)\rangle=|1\rangle,\quad|1\oplus f(x)\rangle=|0\rangle 
\end{aligned}
et donc,
\begin{aligned}
\frac{|1\rangle-|0\rangle}{\sqrt{2}}=-\frac{|0\rangle-|1\rangle}{\sqrt{2}}=-|-\rangle
\end{aligned}


Ainsi, aprs l'application de l'oracle, l'tat des qubits devient :
\begin{align}
        \frac{1}{\sqrt{2^n}} \sum_{x=0}^{2^n-1} (-1)^{s \cdot x} |x\rangle 
    \end{align}


Exemple d'application de l'oracle:
Soit $s=101$, l'oracle appliquera donc deux portes CNOT (une entre le 1er qubit et le qubit auxiliaire et une entre le 3e qubit et le qubit auxilaire).

L'tat superpos
\begin{align}
\frac{1}{\sqrt{8}} (\vert 000 \rangle + \vert 001 \rangle + \vert 010 \rangle + \vert 011 \rangle + \vert 100 \rangle + \vert 101 \rangle + \vert 110 \rangle + \vert 111 \rangle)
\end{align}

devient alors
\begin{align}
\frac{1}{\sqrt{8}} (\vert 000 \rangle - \vert 001 \rangle + \vert 010 \rangle - \vert 011 \rangle - \vert 100 \rangle + \vert 101 \rangle - \vert 110 \rangle + \vert 111 \rangle)
\end{align}



### 4. Transformation Hadamard inverse

Appliquez  nouveau la porte Hadamard  chaque qubit d'entre. Cette opration inverse la transformation Hadamard initiale et "concentre" la superposition sur un seul tat binaire, qui correspond  la chane $ s $.

Aprs l'application de cette transformation, l'tat des qubits d'entre est directement $|s\rangle $.

### 5. Mesure

Mesurez les $n $ qubits d'entre. Le rsultat de la mesure sera la chane binaire $ s $, que vous cherchiez.
